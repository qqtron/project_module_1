```python
import pandas as pd

from sqlalchemy import create_engine

sqlitedb_abs_path = '/vagrant/Data/raw_data_project_m1.db'

conn_str = f'sqlite:///{sqlitedb_abs_path}'

engine = create_engine(conn_str)

pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", engine)

#Call all tables
```

```python
df_personal = pd.read_sql_query("SELECT * FROM personal_info", engine)

df_personal
```

```python
#Cleaning gender

gender = df_personal['gender'].to_list()
clean_gender = []

for item in gender:
    if "fem" in item.lower():
        clean_gender.append("Female")
    else:
        clean_gender.append("Male")
        
clean_gender

```

```python
df_personal['gender'] = clean_gender

df_personal['gender']
```

```python
df_personal
```

```python
age = df_personal['age'].to_list()
clean_age = []

for item in age:
    if 'old' in item:
        clean_age.append(int(item[0:2]) + 4) 
    else:
        clean_age.append(2020 - int(item) + 4)

clean_age
```

```python
# Age clean

df_personal['age'] = clean_age

df_personal
```

```python
#Make joins of three tables

my_first_joins = pd.read_sql_query('''SELECT country_info.uuid, career_info.normalized_job_code, personal_info.gender, country_info.country_code
FROM personal_info
JOIN career_info ON career_info.uuid = personal_info.uuid 
JOIN country_info ON personal_info.uuid = country_info.uuid ;  ''', engine)

my_first_joins
```

```python
#Quantity by country 

my_table = pd.read_sql_query('''SELECT country_info.uuid AS 'Country_uuid', career_info.normalized_job_code, personal_info.gender, country_info.country_code,
count (country_info.uuid) AS 'Quantity'
FROM personal_info
JOIN career_info ON career_info.uuid = personal_info.uuid 
JOIN country_info ON personal_info.uuid = country_info.uuid 
GROUP BY country_info.country_code, personal_info.gender ;  ''', engine)

my_table
```
